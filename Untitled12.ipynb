```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

# Your original Q-learning code
n_states = 6
actions = [0, 1]
q_table = np.zeros((n_states, len(actions)))
q_table[0] = [100, 100]
q_table[5] = [40, 40]
gamma = 0.5
n_iterations = 10

# Propagate the rewards
for _ in range(n_iterations):
    old_q_table = q_table.copy()
    for state in range(1, n_states - 1):
        q_table[state, 0] = gamma * np.max(old_q_table[state - 1])
        q_table[state, 1] = gamma * np.max(old_q_table[state + 1])

class RobotGridVisualizer:
    def __init__(self):
        plt.style.use('seaborn')

    def simulate_path(self, start_state):
        # Get path
        path = simulate_bot(start_state)
        path_str = ' â†’ '.join(map(str, path))

        # Animate robot movement through the path taken
        for state in path:
            # Create a new figure for each state
            fig, ax = plt.subplots(figsize=(12, 4))
            fig.canvas.manager.set_window_title(f'Robot Path from State {start_state} to State {state}')

            # Draw grid cells
            for i in range(n_states):
                color = 'gold' if i in [0, 5] else 'white'
                rect = Rectangle((i, 0), 1, 1, facecolor=color, edgecolor='black')
                ax.add_patch(rect)
                # Add state numbers
                ax.text(i + 0.5, -0.1, f'S{i}', ha='center', va='center')
                # Add Q-values
                if i in [0, 5]:
                    ax.text(i + 0.5, 1.1, f'R={q_table[i][0]}', ha='center', va='center')
                else:
                    ax.text(i + 0.5, 1.1, f'L:{q_table[i][0]:.1f}\nR:{q_table[i][1]:.1f}',
                           ha='center', va='center')

            # Create robot at current state position
            robot = Circle((state + 0.5, 0.5), 0.3, color='blue', alpha=0.7)
            ax.add_patch(robot)

            # Set axis properties
            ax.set_xlim(-0.5, n_states - 0.5)
            ax.set_ylim(-0.5, 1.5)
            ax.axis('off')
            ax.set_title(f'Robot Navigation from State {start_state} to State {state}')

            # Draw and pause to show current state
            plt.draw()
            plt.pause(1.0)   # Pause to visualize movement

        # Keep the window open for a moment before next simulation
        plt.pause(2.0)
        return path

def simulate_bot(start_state):
    state = start_state
    path = [state]

    while state not in [0, n_states - 1]:
        action = np.argmax(q_table[state])
        state = state - 1 if action == 0 else state + 1
        path.append(state)

    return path

# Print Q-table
print("Final Q-table:")
print(q_table)

# Create visualizer and run simulations
visualizer = RobotGridVisualizer()

# Simulate paths from different starting states
for start_state in range(1, n_states-1):
    print(f"\nSimulating robot from state {start_state}")
    path = visualizer.simulate_path(start_state)
    print(f"Path taken: {path}")

plt.show()  # Keep all windows open until manually closed
```
